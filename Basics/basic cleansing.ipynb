```python
import pandas as pd
billboard=pd.read_csv('billboard.csv')
```

```python
billboard.head()
```

```python
billboard.columns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(billboard.loc[0,'x1st.week':'x76th.week'])
```

```python
bshort=billboard[['artist.inverted','track','time','date.entered',
                  'x1st.week','x2nd.week', 'x3rd.week']]
bshort.head()
```

```python
bshort.columns=['artist','track','time','date','wk1','wk2','wk3']
bshort.head()
```

```python
bmelt=bshort.melt(['artist','track','time','date'],['wk1','wk2','wk3'],'week','rank')
bmelt.head()
```

```python
bmelt['week'] = bmelt['week'].apply(lambda s: int(s[2]))
bmelt.head()
```

```python
bmelt['date'] = pd.to_datetime(bmelt['date'])
bmelt.head()
```
