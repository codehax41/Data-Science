```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('hadleydataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
melted = df.melt(['country', 'year'],['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'],'sexage','cases')
```

```python
melted.head()
```

```python
melted['sex'] = melted['sexage'].str.slice(0,1)
melted['age'] =melted['sexage'].str.slice(1)
```

```python
melted.head()
```

```python
melted.drop(['sexage'], axis=1,inplace=True)
```

```python
melted.head()
```

```python
melted['age'] = melted['age'].map({'04':'0-4','514':'5-14','1524':'15-24','2534':'25-34',
                                  '3544':'35-44','4554':'45-54','5564':'55-64','65':'65+','u':np.nan})
```

```python
melted.tail()
```

```python
final=melted.dropna(subset=['cases'])
```

```python
final.tail()
```

```python
final.sort_values(['country', 'year', 'cases','age', 'sex'])
```

```python
final.to_csv('final_hadleyclean.csv', index=False)
```
